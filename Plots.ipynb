```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
x = list(np.random.normal(300,100,500))+ list(np.random.normal(1100,100,500))+list(np.random.normal(2000,200,500))
min(x), max(x)
```

```python
x_ = np.random.normal(np.mean(x), np.std(x), 1500)
```

```python
fig,axs = plt.subplots(1,1, figsize=(15,5))
sns.distplot(x,bins=100,kde=False)
sns.distplot(x_,bins=100, kde=False)
axs.set_xlabel("Listing Price($)")
axs.set_title("Histogram of Digital Type Watch Listing Price($)")
axs.text(200,10,"Mean: {}$".format(300.))
axs.text(200,5,"STD: {}$".format(100.))
axs.text(1000,10,"Mean: {}$".format(1100.))
axs.text(1000,5,"STD: {}$".format(100.))
axs.text(1900,10,"Mean: {}$".format(2000.))
axs.text(1900,5,"STD: {}$".format(200.))

axs.text(2500,50,"Mean: {}$".format(np.round(np.mean(x),2)))
axs.text(2500,45,"Standard Deviation: {}$".format(np.round(np.std(x),2)))
axs.legend(["Underlying Distribution","Emperical Distribution"])
```

```python
xT = np.arange(700,2700,250)
```

```python
x1 = []
y1 = []
for i in xT:
    temp = list(np.random.normal(i,i/10.,100)) + list(np.random.normal(i+500,i/20.,100))
    y1+=temp
    x1+=[i]*200
```

```python
plt.rcParams.update({'font.size': 15})
fig, axs = plt.subplots(1,1, figsize=(12,7))
sns.violinplot(x1,y1, palette="muted", split=True)
axs.set_xticklabels(["Type: {}".format(i) for i in range(1,9)])
axs.set_xlabel("Watch Type")
axs.set_ylabel("Price($)")
axs.set_title("Price distribution by Watch Type")
```
